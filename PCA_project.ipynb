```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
min_max_scaler.fit(df1_list)
```

```python
# get scaled data 
dataset_scaled = pd.DataFrame(min_max_scaler.transform(df1_list))
# assign column labels
dataset_scaled.columns 
```

```python
dataset_scaled.describe()
```

```python
# import PCA from decomposition
from sklearn.decomposition import PCA

# create the PCA object to use to fit the given data
pca = PCA()

# fit the PCA with the given cleaned-up data
pca.fit(dataset_scaled)
```

```python
# Principal axes in feature space, representing the directions of maximum variance in the data.
print(pca.components_)
```

```python
# check the shape
print(pca.components_.shape)
```

```python
# check the type 
print(type(pca.components_))
```

```python
# Percentage of variance explained by each of the selected components
print(pca.explained_variance_)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(final_list, test_size=0.30)
```

```python
# considering only the highly correlated features
X_train = train.drop(['SalePrice'], axis = 1)
y_train = train['SalePrice'] # output of the training data

X_test = test.drop(['SalePrice'], axis = 1)
y_test = test['SalePrice']
```

```python
print('Dimension of X_train dataframe:', X_train.shape)
print('Dimension of y_train dataframe:', y_train.shape)
print('Dimension of X_test dataframe:', X_test.shape)
print('Dimension of y_test dataframe:', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

ln_reg = LinearRegression()
ln_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df_pred = ln_reg.predict(X_test)
print('Mean Absolute Error of linear regression:', mean_absolute_error(df_pred, y_test))
print('Mean Square Error of linear regression:', mean_squared_error(df_pred, y_test))
print('R_Squared score of linear regression:', r2_score(y_test,df_pred))
```
